```python
import pandas as pd
import numpy as np
from chembl_webresource_client.new_client import new_client
```

```python
target = new_client.target
target_query = target.search('TNF')
targets = pd.DataFrame.from_dict(target_query)
targets
```

```python
selected_target = targets.target_chembl_id[0]
selected_target
```

```python
activity = new_client.activity
res = activity.filter(target_chembl_id = selected_target).filter(standard_type = 'IC50')
```

```python
df = pd.DataFrame.from_dict(res)
```

```python
new_df = df.drop(['activity_comment', 
         'activity_id', 
         'activity_properties', 
         'assay_chembl_id',
         'assay_description',
         'bao_endpoint',
         'bao_format',
         'bao_label',
         'data_validity_comment',
         'data_validity_description',
         'document_chembl_id',
         'document_journal',
         'document_year',
         'ligand_efficiency',
         'molecule_pref_name',
         'parent_molecule_chembl_id',
         'pchembl_value',
         'potential_duplicate',
         'qudt_units',
         'record_id',
         'relation',
         'src_id' ,
         'standard_flag',
         'standard_relation',
         'standard_text_value',
         'standard_type',
         'standard_units',
         'standard_upper_value',
         'target_pref_name',
         'target_tax_id',
         'text_value',
         'toid',
         'uo_units',
         'upper_value'], axis =1)
```

```python
new_df = new_df.query("assay_type =='B'")
```

```python
new_df = new_df.dropna()
```

```python
new_df.head()
```

```python
data = new_df.drop(['assay_type','target_chembl_id','target_organism','type','units','value'], axis = 1)
```

```python
data
```

```python
bioactivity_class = []
for i in data.standard_value:
  if float(i) >= 10000:
    bioactivity_class.append("inactive")
  elif float(i) <= 1000:
    bioactivity_class.append("active")
  else:
    bioactivity_class.append("intermediate")
```

```python
data
```

```python
bioactivity_class
```

```python
bioactivity_class_dict = {'bioactivity_class':bioactivity_class}
```

```python
bioactivity_class_dict
```

```python
bioactivity_class = pd.DataFrame(bioactivity_class_dict, columns = ['bioactivity_class'])
```

```python
bioactivity_class
```

```python
big_data = pd.concat([data.reset_index(drop = True), bioactivity_class.reset_index(drop = True)], axis = 1)
```

```python
big_data
```

```python
st_value = list(big_data.standard_value)
```

```python
st_value_2 = []
for item in st_value:
    st_value_2.append(float(item))
```

```python
pIC50 = []
for v in st_value_2:
    molar = v*(10.0**-9)
    pIC50.append(-np.log10(molar))
```

```python
pic_data = pd.DataFrame({'pIC50':pIC50}, big_data.molecule_chembl_id)
```

```python
pic_data
```

```python
data_pIC = pd.merge(big_data, pic_data, on = 'molecule_chembl_id')
```

```python
data_pIC
```

```python
big_data_final = big_data[big_data.bioactivity_class != 'intermediate']
```
